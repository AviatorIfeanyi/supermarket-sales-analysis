```python
import pandas as pd
```

```python
data = pd.read_csv('../01_source_data/supermarket_sales.csv', index_col=0)
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
data.columns
```

```python
data['Year'] = data['Date'].dt.year
```

```python
data['Month Name'] = data['Date'].dt.month_name()
```

```python
data['Day Name'] = data['Date'].dt.day_name()
```

```python
data.to_csv('../03_analysis/supermarket_sales.csv', encoding='utf-8', index_label='Invoice ID')
```
